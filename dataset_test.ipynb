```python
import matplotlib.pyplot as plt
import numpy as np

def show_sample(patch, label, idx):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(patch.permute(1, 2, 0))  # Adjust for your data format
    plt.title("Patch_{}".format(idx))
    plt.subplot(1, 2, 2)
    plt.imshow(label, cmap='gray')
    plt.title("Label_{}".format(idx))
    plt.show()
len = test_dataset.__len__()
for i in range(4):  # Visualize the first 5 samples
    rand = np.random.randint(0, len)
    patch, label = test_dataset[rand]
    show_sample(patch, label, rand)
```
